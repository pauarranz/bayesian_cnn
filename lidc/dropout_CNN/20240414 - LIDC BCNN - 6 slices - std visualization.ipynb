# Load LIDC dataset
### Step 1: Load all nodules as a 3D representation in numpy array & the labels from the nodule name

```python
# Select an even number of slices to extract:
num_slices = 6
# Select nodules dataset location
lidc_path = Path('F:\\master\\manifest-1600709154662\\nodules_16slices')
```

```python
images = []
diagnostics = []
nodules_path = list(lidc_path.iterdir())
for nodule_path in nodules_path:
    images_path = list(nodule_path.iterdir())
    # Reduce the slices to the specified amount:
    images_path = images_path[int(len(images_path)/2)-int(num_slices/2):int(len(images_path)/2)+int(num_slices/2)]
    # Merge all slices of a nodule together --> create 3D representation of the nodule
    image_nodule = []
    for image_path in images_path:
        # Load slice image
        image_slice = np.array(Image.open(image_path)) / 255.0
        # Add it to the nodule
        image_nodule.append(image_slice)
    # convert the 3D nodule to numpy array
    image_nodule = np.array(image_nodule)
    
    diag = int(nodule_path.name.split('_')[5])
    # remove incorrect image shape or diagnostic incogruent
    if image_nodule.shape == (num_slices, 64, 64) and diag != 3:
        # Add the nodule 3D representation to the list of images
        images.append(image_nodule)
        # Save nodule label (diagnostic)
        diagnostics.append(diag)
    elif image_nodule.shape != (num_slices, 64, 64):
        print(f'[ERROR] {nodule_path.name} has shape {image_nodule.shape}, but expected shape is: (16, 64, 64)')
# convert the dataset to numpy array
images = np.array(images)
```

```python
def plot_slices(num_rows, num_columns, width, height, data):
    """Plot a montage of X CT slices"""
    data = np.rot90(np.array(data))
    data = np.transpose(data)
    data = np.reshape(data, (num_rows, num_columns, width, height))
    rows_data, columns_data = data.shape[0], data.shape[1]
    heights = [slc[0].shape[0] for slc in data]
    widths = [slc.shape[1] for slc in data[0]]
    fig_width = 12.0
    fig_height = fig_width * sum(heights) / sum(widths)
    f, axarr = plt.subplots(
        rows_data,
        columns_data,
        figsize=(fig_width, fig_height),
        gridspec_kw={"height_ratios": heights},
    )
    for i in range(rows_data):
        for j in range(columns_data):
            axarr[i, j].imshow(data[i, j], cmap="gray")  # Corrected indexing
            #axarr[i, j].imshow(data[i][j], cmap="gray")
            axarr[i, j].axis("off")
    plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, bottom=0, top=1)
    plt.show()


# Visualize montage of slices.
plot_slices(
    num_rows=2, 
    num_columns=3, 
    width=64, 
    height=64, 
    data=images[0].transpose()
)
```

# Data augmentation

Define function to rotate images

```python
def rotate_image(image):
    # define some rotation angles
    angles = [-20, -10, -5, 5, 10, 20]
    # pick angles at random
    angle = random.choice(angles)
    # rotate volume
    image = rotate(image.transpose(), angle).transpose()
    return image
```

Generate new images

```python
images_aug = list(images)
labels_aug = list(labels)
for i, l in zip(images, labels):
    if l==1:
        for rep in range(3):
            images_aug.append(rotate_image(i))
            labels_aug.append(l)
    elif l==0:
        if 0.8 > random.random():
            images_aug.append(rotate_image(i))
            labels_aug.append(l)  

images_aug = np.array(images_aug)
labels_aug = np.array(labels_aug)
```

```python
count = Counter(labels_aug)
df = pandas.DataFrame.from_dict(count, orient='index')
df.plot(kind='bar')
```

Divide between train and test data

```python
# Create train and test dataset
train_images, test_images, train_labels, test_labels = train_test_split(images_aug, labels_aug, test_size=0.33, random_state=42)
```

# Building the CNN:
Models used by others: https://github.com/marichka-dobko/Lung_Nodule_Classification/blob/master/src_keras/models_keras.py

A simple CNN with 5 convolutional layers

```python
def le_net_3d():
    model = keras.Sequential()
    model.add(layers.Convolution3D(6, kernel_size=(2, 5, 5), activation='relu', padding='same', input_shape=(6,64,64,1)))
    model.add(layers.MaxPooling3D(strides=2)) 
    model.add(layers.Convolution3D(16, kernel_size=(2, 5, 5), activation='relu', padding='same'))
    model.add(layers.MaxPooling3D(strides=2)) 

    model.add(layers.Flatten())
    model.add(layers.Dense(120, activation='relu'))
    model.add(layers.Dense(84, activation='relu'))
    model.add(layers.Dense(2, activation='softmax'))

    return model


# Build model.
model = le_net_3d()
model.summary()
```

```python
def fit_model(model, epochs):  
    model.compile(  
        optimizer="adam",  
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),  
        metrics=["accuracy"],  
    )  

    history = model.fit(  
        train_images,  
        train_labels,  
        epochs=epochs,  
        validation_data=(test_images, test_labels),  
    )
    
    # Plot training history
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()
    # Plot final model accuracy
    print(f'Training accuracy: {history.history["accuracy"][-1]}')
    print(f'Validation accuracy: {history.history["val_accuracy"][-1]}')
    return model  


model = fit_model(model, epochs=30)
```

# Adding Dropout

Let’s see what a slightly different model would predict on these images. We’ll now use MC dropout as our model. By sampling, we should be able to increase the models’ uncertainty compared to a standard NN. 

Let’s first define our model:

```python
def le_net_3d_dropout():
    model = keras.Sequential()
    model.add(layers.Convolution3D(6, kernel_size=(2, 5, 5), activation='relu', padding='same', input_shape=(6,64,64,1)))
    
    model.add(layers.Dropout(0.2))
    model.add(layers.MaxPooling3D(strides=2)) 
    model.add(layers.Convolution3D(16, kernel_size=(2, 5, 5), activation='relu', padding='same'))
    model.add(layers.MaxPooling3D(strides=2))
    
    model.add(layers.Dropout(0.5))
    model.add(layers.Flatten())
    model.add(layers.Dense(120, activation='relu'))
    
    #model.add(layers.Dropout(0.5))
    model.add(layers.Dense(84, activation='relu'))
    
    #model.add(layers.Dropout(0.5))
    model.add(layers.Dense(2, activation='softmax'))

    return model
```

Then let’s instantiate it:

```python
dropout_model = le_net_3d_dropout()
dropout_model.summary()
```

```python
dropout_model = fit_model(dropout_model, epochs=30)
```

Our model with dropout will achieve a similar accuracy as our vanilla model. Let’s now run inference with dropout and plot the mean confidence score of MC dropout:

```python
def predictions_dropout(imgs):
    predictions = np.array(  
        [  
            tf.nn.softmax(dropout_model(np.expand_dims(imgs, -1), training=True), axis=1)  
            for _ in range(100)  
        ]  
    )  
    predictions_mean = np.mean(predictions, axis=0)
    return predictions_mean
```

# Define models evaluation method

```python
class ModelEvaluation:
    def __init__(self, model, test_images, test_labels, dropout_inference=False):
        self.pred_prob_dict = {'correct':[], 'incorrect': []}
        self.pred_std_dict = {'correct':[], 'incorrect': []}
        self.pred_prob_std_dict = {'correct':[], 'incorrect': []}
        self.model_accuracy = None

        for test_image, test_label in zip(test_images, test_labels):
            if dropout_inference:
                self.update_dropout_pred_dict(image=test_image, label=test_label, model=model)
            else:
                self.update_classic_pred_dict(image=test_image, label=test_label, model=model)

    def update_dropout_pred_dict(self, model, image, label):
        
        # Simulate Bayesian CNN with dropout at inference (100 test runs)
        pred_list = [  
            dropout_model(np.expand_dims([image], -1), training=True) 
            for _ in range(100)  
        ]
        pred = np.mean(np.array(pred_list), axis=0)
        
        # Compute standard deviation of prediction 
        pred_std_list = [  
            np.argmax(np.array(p[0]))
            for p in pred_list 
        ]
        pred_std = np.mean(np.array(pred_std_list), dtype=np.float64)
        
        # Compute standard deviation of prediction confidence
        pred_prob_std_list = [  
            np.array(p[0]).max() * 100
            for p in pred_list 
        ]
        pred_prob_std = np.mean(np.std(np.array(pred_prob_std_list), dtype=np.float64))
        
        
        # Compute the top predicted class probability
        pred_conf = pred[0].max() 
        pred_conf = int(round(pred_conf*100, 2))
        # Compute the top predicted class
        pred_label = np.argmax(pred[0])
        # Check if the prediction is correct
        pred_correct = pred_label==label
        # Update dictionary
        self.update_pred_dict(pred_correct, pred_conf)
        self.update_pred_std_dict(pred_correct, pred_std)
        self.update_pred_prob_std_dict(pred_correct, pred_prob_std)
    
    def update_classic_pred_dict(self, model, image, label):
        
        pred = model.predict(np.expand_dims([image], -1), verbose=0)
        # Get the index of the class with the highest probability
        pred_label = np.argmax(pred)
        # Compute the top predicted class probability
        pred_conf = pred[0][pred_label]
        pred_conf = int(round(pred_conf*100, 2))
        
        # Check if the prediction is correct
        pred_correct = pred_label==label
        
        self.update_pred_dict(pred_correct, pred_conf)
    
    def update_pred_dict(self, pred_correct, pred_conf):
        # Update dictionary
        if pred_correct:
            self.pred_prob_dict['correct'].append(pred_conf)
        else:
            self.pred_prob_dict['incorrect'].append(pred_conf)
            
    def update_pred_std_dict(self, pred_correct, pred_std):
        # Update dictionary
        if pred_correct:
            self.pred_std_dict['correct'].append(pred_std)
        else:
            self.pred_std_dict['incorrect'].append(pred_std)
            
    def update_pred_prob_std_dict(self, pred_correct, pred_prob_std):
        # Update dictionary
        if pred_correct:
            self.pred_prob_std_dict['correct'].append(pred_prob_std)
        else:
            self.pred_prob_std_dict['incorrect'].append(pred_prob_std)
    
    def plot_pred_dist(self, num_bins=25):
        plt.figure(figsize=(30, 10))
        # Plot probability histogram
        bins = np.linspace(50, 100, num_bins)

        # Correct predictions probability distribution
        plt.hist(
            self.pred_prob_dict['correct'],
            bins,
            weights=np.ones(len(self.pred_prob_dict['correct'])) / len(self.pred_prob_dict['correct']),
            alpha=0.5,
            label='correct',
            color='g',
        )
        # Incorrect predictions probability distribution
        plt.hist(
            self.pred_prob_dict['incorrect'],
            bins,
            weights=np.ones(len(self.pred_prob_dict['incorrect'])) / len(self.pred_prob_dict['incorrect']),
            alpha=0.5,
            label='incorrect',
            color='r',
        )
        # Add correct predictions mean probability
        plt.axvline(x=np.mean(self.pred_prob_dict['correct']), color='g', label='Correct predictions - mean probability')
        # Add incorrect predictions mean probability
        plt.axvline(x=np.mean(self.pred_prob_dict['incorrect']), color='r', label='Incorrect predictions - mean probability')
        # Add model accuracy
        plt.axvline(x=self.get_model_accuracy(), color='b', label='Model accuracy')
        # Format y axis as a percentage (example: from 0.8 to 80%)
        plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
        # Add legend
        plt.legend(loc='upper right')
        plt.title('Prediction probability distribution')
        plt.xlabel('Prediction probability (%)')
        plt.ylabel('Distribution')
        plt.show()
        plt.close()
        
    def plot_pred_prob_std_dist(self, max_value=20, num_bins=100):
        plt.figure(figsize=(30, 10))
        # Plot probability histogram
        bins = np.linspace(0, max_value, num_bins)
        # Correct predictions probability distribution
        plt.hist(
            self.pred_prob_std_dict['correct'],
            weights=np.ones(len(self.pred_prob_std_dict['correct'])) / len(self.pred_prob_std_dict['correct']),
            alpha=0.5,
            label='correct',
            color='g',
            bins=bins,
        )
        # Incorrect predictions probability distribution
        plt.hist(
            self.pred_prob_std_dict['incorrect'],
            weights=np.ones(len(self.pred_prob_std_dict['incorrect'])) / len(self.pred_prob_std_dict['incorrect']),
            alpha=0.5,
            label='incorrect',
            color='r',
            bins=bins,
        )
        # Add correct predictions mean probability
        plt.axvline(x=np.mean(self.pred_prob_std_dict['correct']), color='g', label='Correct predictions - prediction probability std')
        # Add incorrect predictions mean probability
        plt.axvline(x=np.mean(self.pred_prob_std_dict['incorrect']), color='r', label='Incorrect predictions - prediction probability std')
        # Format y axis as a percentage (example: from 0.8 to 80%)
        plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
        # Add legend
        plt.legend(loc='upper right')
        plt.title('Prediction probability standard deviation distribution')
        plt.xlabel('Prediction probability standard deviation')
        plt.ylabel('Distribution')
        plt.show()
        plt.close()
    
    def plot_pred_std_dist(self, num_bins=100):
        plt.figure(figsize=(30, 10))
        # Plot probability histogram
        bins = np.linspace(0, 1, num_bins)
        # Correct predictions probability distribution
        plt.hist(
            self.pred_std_dict['correct'],
            weights=np.ones(len(self.pred_std_dict['correct'])) / len(self.pred_std_dict['correct']),
            alpha=0.5,
            label='correct',
            color='g',
            bins=bins,
        )
        # Incorrect predictions probability distribution
        plt.hist(
            self.pred_std_dict['incorrect'],
            weights=np.ones(len(self.pred_std_dict['incorrect'])) / len(self.pred_std_dict['incorrect']),
            alpha=0.5,
            label='incorrect',
            color='r',
            bins=bins,
        )
        # Add correct predictions mean probability
        plt.axvline(x=np.mean(self.pred_std_dict['correct']), color='g', label='Correct predictions - prediction label std')
        # Add incorrect predictions mean probability
        plt.axvline(x=np.mean(self.pred_std_dict['incorrect']), color='r', label='Incorrect predictions - prediction label std')
        # Format y axis as a percentage (example: from 0.8 to 80%)
        plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
        # Add legend
        plt.legend(loc='upper right')
        plt.title('Prediction standard deviation distribution')
        plt.xlabel('Prediction label standard deviation')
        plt.ylabel('Distribution')
        plt.show()
        plt.close()
        

    def get_pred_prob_dict(self):
        return self.pred_prob_dict
    
    def get_pred_prob_std_dict(self):
        return self.pred_prob_std_dict
    
    def get_pred_std_dict(self):
        return self.pred_std_dict
    
    def get_model_accuracy(self):
        num_pred_co = len(self.pred_prob_dict['correct'])
        num_pred_in = len(self.pred_prob_dict['incorrect'])
        return (num_pred_co / (num_pred_co + num_pred_in)) * 100
```

## Improving bayesian model prediction probability (explainability)

Observing the prediction probability distribution, is concluded that correct predictions are correctly labeled with high probability. Incorrect predictions on the other side have a much worst distribution.

Observing the prediction probability standard distribution, the inverse is observed: correct predictions distribution is not accurate, but incorrect predictions distributions is much more accurate.

Therefore, both results will be merged to create a new bayesian prediction probability.

```python
pred_prob_dict = me_bayesian.get_pred_prob_dict()
pred_std_dict = me_bayesian.get_pred_std_dict()
pred_prob_std_dict = me_bayesian.get_pred_prob_std_dict()

pred_prob_weighted_dict = {}

# The model prediction probability is weighted with the model standard deviation
for label in ['correct', 'incorrect']:
    pred_prob_weighted_dict[label] = np.array(pred_prob_dict[label]) - (np.array(pred_prob_std_dict[label]) - 7)

# Shift min value to 0
for label in ['correct', 'incorrect']:
    min_correct = np.array(pred_prob_weighted_dict['correct']).min()
    min_incorrect = np.array(pred_prob_weighted_dict['incorrect']).min()
    offset = np.array(min_correct, min_incorrect).min()
    
    max_correct = np.array(pred_prob_weighted_dict['correct']).max() - offset
    max_incorrect = np.array(pred_prob_weighted_dict['incorrect']).max() - offset
    range_values = np.array(max_correct, max_incorrect).max()

    # Change from 0-X% to 0-100%    
for label in ['correct', 'incorrect']:
    pred_prob_weighted_dict[label] = ((np.array(pred_prob_weighted_dict[label])-offset)/range_values)*100
```

Plot bayesian prediction probability

```python
def plot_bayesian_pred_prob_dist(pred_prob_dict):
    plt.figure(figsize=(30, 10))
    # Plot probability histogram
    bins = np.linspace(0, 100, 25)
    # Correct predictions probability distribution
    plt.hist(
        pred_prob_dict['correct'],
        weights=np.ones(len(pred_prob_dict['correct'])) / len(pred_prob_dict['correct']),
        alpha=0.5,
        label='correct',
        color='g',
        bins=bins,
    )
    # Incorrect predictions probability distribution
    plt.hist(
        pred_prob_dict['incorrect'],
        weights=np.ones(len(pred_prob_dict['incorrect'])) / len(pred_prob_dict['incorrect']),
        alpha=0.5,
        label='incorrect',
        color='r',
        bins=bins,
    )
    # Add correct predictions mean probability
    plt.axvline(x=np.mean(pred_prob_dict['correct']), color='g', label='Correct predictions - prediction label std')
    # Add incorrect predictions mean probability
    plt.axvline(x=np.mean(pred_prob_dict['incorrect']), color='r', label='Incorrect predictions - prediction label std')
    # Format y axis as a percentage (example: from 0.8 to 80%)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    # Add legend
    plt.legend(loc='upper right')
    plt.title('Bayesian prediction correctness distribution')
    plt.xlabel('Prediction probability, weighted with prediction probability standard deviation')
    plt.ylabel('Distribution')
    plt.show()
    plt.close()

    print(f'incorrect mean: {np.mean(pred_prob_dict["incorrect"])}')
    print(f'correct mean: {np.mean(pred_prob_dict["correct"])}')
```

Seeing this result, the next question is obvious:

**Can the model accuracy be improved just by modifying the final judgement on this binary classification?**

Let's try it out!

The first step is to create a new dictionary with this "improved" clasification, by modifying the judgement of the low probability samples, to do so the threshold is defined with the incorrect samples mean probability:

```python
pred_prob_weighted_dict_improved = {"incorrect": [], "correct": []}
threshold = 37.98 # Defined as the incorrect mean

pred_prob_weighted_dict_improved['correct'].extend(
    list(pred_prob_weighted_dict["incorrect"][pred_prob_weighted_dict["incorrect"] < threshold])
)
pred_prob_weighted_dict_improved['correct'].extend(
    list(pred_prob_weighted_dict["correct"][pred_prob_weighted_dict["correct"] >= threshold])
)

pred_prob_weighted_dict_improved['incorrect'].extend(
    list(pred_prob_weighted_dict["correct"][pred_prob_weighted_dict["correct"] < threshold])
)
pred_prob_weighted_dict_improved['incorrect'].extend(
    list(pred_prob_weighted_dict["incorrect"][pred_prob_weighted_dict["incorrect"] >= threshold])
)
```

Visualize the improved model accuracy:

```python
print(f'Original bayesian model accuracy: {me_bayesian.get_model_accuracy()}')

num_pred_co = len(pred_prob_weighted_dict_improved['correct'])
num_pred_in = len(pred_prob_weighted_dict_improved['incorrect'])
model_acc_impv = (num_pred_co / (num_pred_co + num_pred_in)) * 100

print(f'Improved bayesian model accuracy: {model_acc_impv}')
```

It can be observed that the accuracy has slightly been reduced, why?

```python
print(f'len(pred_prob_weighted_dict["correct"] =  {len(pred_prob_weighted_dict["correct"])}')
print(f'len(pred_prob_weighted_dict["incorrect"]) = {len(pred_prob_weighted_dict["incorrect"])}')

print(f'len(pred_prob_weighted_dict_improved["correct"] =  {len(pred_prob_weighted_dict_improved["correct"])}')
print(f'len(pred_prob_weighted_dict_improved["incorrect"]) = {len(pred_prob_weighted_dict_improved["incorrect"])}')
```

Observing the number of correct and incorrect predictions, it can be observed that since the model already had a high accuracy, the number of correct predictions is much higer than the number of incorrect ones.

Therefore, while it's true that percentually there are much more incorrect predictions in the low probability area (changed area), in absolute terms the number of correct predictions is slightly bigger in the low probability area.

# Conclusions
The bayesian model allowed to improve the model predictions explainability, by combining:

    - The model prediction probability.
    - The model prediction probability standard deviation.

Although, it needs to be highlighted that this explainability improvement cannot be extrapolated to an accuracy improvement.
